```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline

sns.set_style("darkgrid")
sns.set_palette("colorblind")
sns.set_context("paper")
```

```python
df = pd.read_csv('xy_match_average.csv')
df
```

```python
fig, ax = plt.subplots()
# set ylabel
ax.set_ylabel('y')
# set xlabel
ax.set_xlabel('x')
# xlim, ylim
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)

plt.scatter(df['x'], df['y'], s=0.05)
plt.show()
```

```python
p = sns.regplot(x=df['x'], y=df['y'], data=df, )
```

```python
line = p.get_children()[1]
#get equation from line
points = line.get_xydata()
points
# get the slope and intercept
slope = (points[99,1] - points[0,1]) / (points[99,0] - points[0,0])
intercept = points[0,1] - (slope * points[0,0])
print("Line: y = {:.6f}x + {:.6f}".format(slope, intercept))
```
