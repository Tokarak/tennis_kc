```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline

sns.set_style("darkgrid")
sns.set_palette("colorblind")
sns.set_context("paper")
```

```python
df = pd.read_csv('xy_match_average.csv')
df
```

```python
fig, ax = plt.subplots()
# set ylabel
ax.set_ylabel('y')
# set xlabel
ax.set_xlabel('x')
# xlim, ylim
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)

plt.scatter(df['x'], df['y'], s=0.05)
plt.show()
```

```python
sns.displot(df['x'], kde=False, bins=50)
```

```python
sns.displot(df['y'], kde=False, bins=30)
```

```python
sns.jointplot(x=df['x'], y=df['y'], kind='hex')
```
